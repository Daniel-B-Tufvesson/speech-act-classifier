```python
from context import speechact
import speechact.corpus as corp
import speechact.annotate as anno
import speechact.corpstats as st
import speechact.classifier.embedding as emb

#labels = anno.SpeechActLabels.get_labels()
labels = [str(act) for act in emb.SPEECH_ACTS]
```

```python
import matplotlib.pyplot as plt

def plot_class_frequencies(corpus: corp.Corpus, labels: list[str]):
    frequencies = st.speech_act_frequencies(corpus)
    counts = [frequencies[label] for label in labels]

    for label in labels:
        print(f'{label}: {frequencies[label]}')

    # Plot the class frequencies.
    plt.figure(facecolor='white')
    plt.title("Number of sentences per class")
    plt.bar(x = labels, height = counts)
    plt.xticks(rotation=90)
    plt.show()

def print_and_plot_data(corpus: corp.Corpus):
    print(corpus.name)
    print(f'Sentences: {corpus.sentence_count}')
    print()

    # Print the class frequencies.
    print('Corpus data:')
    plot_class_frequencies(corpus, labels)
```

```python
# Load the data.
test_set = corp.Corpus('../data/annotated data/test-set.conllu.bz2')
test_set_upsampled = corp.Corpus('../data/annotated data/test-set-upsampled.conllu.bz2')
dev_set = corp.Corpus('../data/annotated data/dev-set.conllu.bz2')
dev_test_set = corp.Corpus('../data/annotated data/dev-test-set.conllu.bz2')
dev_train_set = corp.Corpus('../data/annotated data/dev-train-set.conllu.bz2')
dev_test_set_upsampled = corp.Corpus('../data/annotated data/dev-test-set-upsampled.conllu.bz2')
dev_train_set_upsampled = corp.Corpus('../data/annotated data/dev-train-set-upsampled.conllu.bz2')
train_set = corp.Corpus('../data/auto-annotated data/train-set.conllu.bz2')
```
