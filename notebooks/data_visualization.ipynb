```python
from context import speechact
import speechact.corpus as corp
import speechact.annotate as anno
import speechact.corpstats as st
import pprint
```

```python
corpus = corp.Corpus('../data/annotated data/dev-set-sentiment.conllu.bz2')

print(f'Sentences: {corpus.sentence_count}')
print()

# Get class frequencies for corpus.
labels = anno.SpeechActLabels.get_labels()
corpus_frequencies = st.speech_act_frequencies(corpus)
corpus_counts = [corpus_frequencies[label] for label in labels]

# Print the class frequencies.
print('Corpus data:')
for label in labels:
    print(f'{label}: {corpus_frequencies[label]}')

# Plot the class frequencies.
import matplotlib.pyplot as plt
plt.figure(facecolor='white')
plt.title("Number of sentences per class: training data")
plt.bar(x = labels, height = corpus_counts)
plt.xticks(rotation=90)
plt.show()
```

```python
test_corpus = corp.Corpus('../data/annotated data/dev-set-sentiment-test.conllu.bz2')
train_corpus = corp.Corpus('../data/annotated data/dev-set-sentiment-train.conllu.bz2')

# Print sentence counts.
print(f'Test sentences: {test_corpus.sentence_count}')
print(f'Training sentences: {train_corpus.sentence_count}')
print()

# Get the class frequencies in training data for each label.
labels = anno.SpeechActLabels.get_labels()
train_frequencies = st.speech_act_frequencies(train_corpus)
train_counts = [train_frequencies[label] for label in labels]

# Print the class frequencies.
print('Training data:')
for label in labels:
    print(f'{label}: {train_frequencies[label]}')

# Plot the class frequencies.
import matplotlib.pyplot as plt
plt.figure(facecolor='white')
plt.title("Number of sentences per class: training data")
plt.bar(x = labels, height = train_counts)
plt.xticks(rotation=90)
plt.show()

# Get the class frequencies in test data for each label.
test_frequencies = st.speech_act_frequencies(test_corpus)
test_counts = [test_frequencies[label] for label in labels]

# Print the class frequencies.
print('Test data: ')
for label in labels:
    print(f'{label}: {test_frequencies[label]}')

# Plot the class frequencies.
plt.figure(facecolor='white')
plt.title("Number of sentences per class: test data")
plt.bar(x = labels, height = test_counts)
plt.xticks(rotation=90)
plt.show()
```
