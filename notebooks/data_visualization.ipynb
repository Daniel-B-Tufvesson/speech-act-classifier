```python
from context import speechact
import speechact.corpus as corp
import speechact.annotate as anno
import speechact.corpstats as st

labels = anno.SpeechActLabels.get_labels()
```

```python
import matplotlib.pyplot as plt

def plot_class_frequencies(corpus: corp.Corpus, labels: list[str]):
    frequencies = st.speech_act_frequencies(corpus)
    counts = [frequencies[label] for label in labels]

    for label in labels:
        print(f'{label}: {frequencies[label]}')

    # Plot the class frequencies.
    plt.figure(facecolor='white')
    plt.title("Number of sentences per class")
    plt.bar(x = labels, height = counts)
    plt.xticks(rotation=90)
    plt.show()
```

```python
corpus = corp.Corpus('../data/annotated data/dev-set-sentiment.conllu.bz2')

print(f'Sentences: {corpus.sentence_count}')
print()

# Print the class frequencies.
print('Corpus data:')
plot_class_frequencies(corpus, labels)
```

```python
test_corpus = corp.Corpus('../data/annotated data/dev-set-sentiment-test.conllu.bz2')
train_corpus = corp.Corpus('../data/annotated data/dev-set-sentiment-train.conllu.bz2')

# Print sentence counts.
print(f'Test sentences: {test_corpus.sentence_count}')
print(f'Training sentences: {train_corpus.sentence_count}')
print()

# Print the class frequencies.
print('Training data:')
plot_class_frequencies(train_corpus, labels)

# Print the class frequencies.
print('Test data: ')
plot_class_frequencies(test_corpus, labels)
```
